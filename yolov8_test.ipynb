```python
import cv2

def drawdot_img(coordinates):
    # è¯»å–äº†ä¸€å¼ å›¾åƒåˆ° img ä¸­
    img = cv2.imread('/home/bdhapp/shuli/dataset/images/train/31.jpg')
 
    color = (256,0,0)  # ä½¿ç”¨è“è‰²
    thickness = -1  # æ¡†çš„åŽšåº¦ï¼Œ-1ä¸ºå¡«å……
    # é€ä¸ªç»˜åˆ¶
    for coord in coordinates:
        x, y, x2, y2 = int(coord["x_min"]), int(coord["y_min"]), int(coord["x_max"]), int(coord["y_max"])
              
        # åœ†å¿ƒåæ ‡
        cx, cy = (x + x2) // 2, (y + y2) // 2
        
        # ç»˜åˆ¶å®žå¿ƒåœ†
        cv2.circle(img, center=(cx,cy), radius=5, color=color, thickness=thickness)
    # æ˜¾ç¤ºå›¾åƒ
    output_path = '/home/bdhapp/shuli/runs/detect/predict4/31_circle.jpg'
    cv2.imwrite(output_path, img)
drawdot_img(coord)
```

```python
import os
path = "/home/bdhapp/shuli/datasets/test_dataset/labels/train"
out_path = "/home/bdhapp/shuli/datasets/test_dataset/labels/test"
for file_name in os.listdir(path):
    file_path = os.path.join(path, file_name)
    write_path = os.path.join(out_path, file_name)
    l = []
    l_w = []
    with open(file_path, 'r', encoding='utf-8') as f:
        l = f.readlines()
    for line in l:
        tmp = line.split(' ')
        tmp[0] = '0'
        l_w.append(' '.join(tmp))
    with open(write_path, 'a+', encoding='utf-8') as f:
        for w_line in l_w:
            f.write(w_line)
```
